```python
#Import python packages
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Splitting a Dataset Into Training and Test Sets

"""
train: [0-2021]
test: [2022-X]

"""
print("Working")
```

```python
#Load Data
df = pd.read_csv("BTC-USD.csv")
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
#Removing Null Values
df = df.dropna()
df
```

```python
df.isna().sum()
```

```python
df.describe()
```

```python
#Middle Value between Trainining/Test Sets
index = list(df.index.values)  
```

```python
for i in index:
    if df["Date"][i].split("-")[0] == "2022":
        imp_value = i
        break
        
print(imp_value)
```

```python
#Train/Test


training_set = df[:imp_value]
test_set = df[imp_value:]
```

```python
training_set
```

```python
test_set
```

```python
training_set = training_set.iloc[:,1:2]
training_set.head()
```

```python
training_set = training_set.values
training_set
```

```python
# Feature Scaling
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
training_set = sc.fit_transform(training_set)
training_set
```

```python
len(training_set)
```

```python
# Getting the inputs and the ouputs
X_train = training_set[0:len(training_set)-1]
y_train = training_set[1:len(training_set)]

# Sample

t_x_0 = pd.DataFrame(X_train[0:5])
t_x_1= pd.DataFrame(y_train[0:5])
sa = pd.concat([t_x_0, t_x_1], axis=1)
sa.columns = (['t_x_0', 't_x_1'])
sa

```

```python
# Reshaping into required shape for Keras
X_train = np.reshape(X_train, (imp_value-1, 1, 1))
X_train
```

```python
# Importing the Keras libraries and packages
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM

"""
What Are LSTMs? LSTM (Long Short-Term Memory) is a Recurrent Neural Network (RNN) based architecture that is widely used in natural language processing and time series forecasting.
Using a series of 'gates,' each with its own RNN, the LSTM manages to keep, forget or ignore data points based on a probabilistic model


"""

#A recurrent neural network (RNN) is a type of artificial neural network which uses sequential data or time series data. 
print("Working")
```

```python
# Initializing the Recurrent Neural Network
model = Sequential()
```

```python
# Adding the input layer and the LSTM layer


#(conda install numpy=1.19.5 --> RESTART THE KERNEL)
model.add(LSTM(units = 4, activation = 'sigmoid', input_shape = (None, 1)))
```

```python
# Adding the output layer
model.add(Dense(units = 1))
```

```python
# Compiling the RNN
model.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
# Fitting the RNN to the Training set
model.fit(X_train, y_train, batch_size = 32, epochs = 200)
```

```python
#TESTING 1
```

```python
test_set.head()
```

```python
test_set.tail()
```

```python
# Getting relevant feature
real_stock_price = (test_set.iloc[:,1:2])
real_stock_price.head()
```

```python
real_stock_price.tail()
```

```python
# Converting to 2D array
real_stock_price = real_stock_price.values
```

```python
# Getting the predicted stock price of 2022

#18 VALUES
inputs = real_stock_price
inputs = sc.transform(inputs)
inputs = np.reshape(inputs, (len(test_set), 1, 1))
predicted_stock_price = model.predict(inputs)
predicted_stock_price = sc.inverse_transform(predicted_stock_price)
```

```python
print(predicted_stock_price)
```

```python
# Visualizing the results
plt.plot(real_stock_price, color = 'green', label = 'Real Stock Price')
plt.plot(predicted_stock_price, color = 'red', label = 'Predicted Stock Price')
plt.title('Stock Price Prediction')
plt.xlabel('Days')
plt.ylabel('Stock Price')
plt.legend()
plt.show()
```
